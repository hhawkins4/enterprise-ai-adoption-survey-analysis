# Reorganizing Data Structure

```python
#Reloading dataset so row slicing always happens on original data
labels_df = pd.read_excel("FILE A.xlsx")

#Removing first two rows: Qualtrics metadata (question text & import IDs)
#Resetting index so pandas can pick up responses 1-4
labels_df = labels_df.iloc[2:].reset_index(drop=True)

#Checking dataset
labels_df.head()
```

```python
#Checking for duplicates (I noticed in File A there were no Response ID duplicates,
#so I am checking if two rows have identical answers across all questions even with different Response #
#(Example: one response got recorded twice for some reason)
labels_no_id = labels_df.drop(columns=['ExternalReference'])
duplicate_responses = labels_no_id.duplicated().sum()

print("# of duplicates:", duplicate_responses)
#Looks like we're good
```

```python
#Checking invalid responses "Prefer not to answer" (I didn't see any, but just in case I missed it)
labels_df.apply(lambda col: col.astype(str).str.contains("Prefer", case=False).sum())

#Looks like there are none
```

```python
#Standardizing all blank spaces, just to make sure all white spaces/no responses are
#counted as missing values
labels_df = labels_df.replace(r'^\s*$', np.nan, regex=True)
```

```python
#Now separating question columns for percentage computation
question_columns = labels_df.columns.drop("ExternalReference")
```

# Grouping questions into single select, multi select, and text fields.

```python
import re
from collections import defaultdict

#Grouping multi-select columns
multi_select_groups = defaultdict(list)
single_select_columns = []

#Identifying columns that follow the Q(x)_(y) format
for col in question_columns:
    if re.match(r"^Q\d+_\d+$", col):
        base_question = col.rsplit("_", 1)[0]
        multi_select_groups[base_question].append(col)
    #Excluding TEXT columns
    elif not col.endswith("_TEXT"):
        single_select_columns.append(col)

#Sorting multi-select columns numerically within each question group
for question in multi_select_groups:
    multi_select_groups[question] = sorted(
        multi_select_groups[question],
        key=lambda x: int(x.split("_")[1])
    )

#Creating summary results
results = []
```

```python
#Creating counts for single-select questions
for col in single_select_columns:
    valid_responses = labels_df[col].dropna()

    #Defining count
    counts = valid_responses.value_counts()
    total_valid = counts.sum()

    for response, count in counts.items():
        #Calculating percent
        percent = (count / total_valid) * 100 if total_valid > 0 else 0

        results.append({
            "question": col,
            "response_option": response,
            "count_n": count,
            "percent": round(percent, 2),
            "denominator_n": total_valid,
            "question_type": "single_select"
        })
```

```python
#Creating counts for multi-select questions
#For the denominator, I used the number of respondents who selected at least one option
for question, cols in multi_select_groups.items():
    subset = labels_df[cols]

    #Number of respondents who selected at least one option
    respondents_answered = subset.notna().any(axis=1).sum()

    for col in cols:
        count = subset[col].notna().sum()

        #I used the actual answer text from the labels file instead of Q10_1, Q10_2, etc.
        non_missing_values = subset[col].dropna().unique()
        if len(non_missing_values) == 1:
            option_label = non_missing_values[0]
        elif len(non_missing_values) > 1:
            option_label = " / ".join(sorted(map(str, non_missing_values)))
        else:
            option_label = col

        #Calculating Percent
        percent = (count / respondents_answered) * 100 if respondents_answered > 0 else 0

        results.append({
            "question": question,
            "response_option": option_label,
            "count_n": count,
            "percent": round(percent, 1),
            "denominator_n": respondents_answered,
            "question_type": "multi_select"
        })
```

```python
#Results table
results_df = pd.DataFrame(results)

#Cleaning up table:

#Sorting questions numerically so Q10 appears after Q9 instead of before Q2
results_df["question_number"] = results_df["question"].str.extract(r"Q(\d+)").astype(int)
#Fixing "Äì" error that shows up in some answers, small nitpick
results_df["response_option"] = results_df["response_option"].str.replace("‚Äì", "–")
#Rounding percentage
results_df["percent"] = results_df["percent"].round(1)

#Creating final table, sorting how I want for each column
results_df = results_df.sort_values(
    by=["question_number", "question_type", "count_n", "response_option"],
    ascending=[True, True, False, True]
).drop(columns="question_number").reset_index(drop=True)

print(results_df.to_string(index=False))
```

```python
#Exporting this cleaned dataset into excel for future data analysis or creating visualizations
results_df.to_excel("survey_summary_results.xlsx", index=False)
```
